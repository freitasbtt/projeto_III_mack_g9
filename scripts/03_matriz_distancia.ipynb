```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
K = 3
```

```python
matriz_scores = pd.read_csv('../datasets/matriz_produto_usuario.csv')
```

```python
matriz_scores
```

```python
# Extrai os id_produto como um array
# Será usado para nomear linhas e colunas da matriz de similaridade
ids_produtos = matriz_scores['id_produto'].values

# Converte o dataframe, sem acoluna de id_produto, em um array numpy
X_scores = matriz_scores.drop(columns='id_produto').to_numpy(dtype=float)

# Cálculo da similaridade cosseno
# O resultado é uma matriz quadrada, com cada valor remetendo a 2 produtos
# As dimensões para calcular as "distâncias" são os usuários
    # Se houvesse apenas 3 usuários, seria possível visualizar cada produto como um ponto em um cubo, com sua posição determinada pelo score de cada um desses três usuários
    # Mas há 1000 usuários, então isso não é visualizável da mesma forma
similaridade_cosseno_produtos = cosine_similarity(X_scores)

# Retornando o resultado para um dataframe legível
similaridade_item_item = pd.DataFrame(
    similaridade_cosseno_produtos,
    index=ids_produtos,
    columns=ids_produtos
)

# Arredondamento para fins de apresentação e eventual poupança de memória
similaridade_item_item = similaridade_item_item.round(6)
```

```python
similaridade_item_item.to_csv('../dados_sinteticos/dados_tratatos/similaridade_item_item.csv')
```

```python
similaridade_item_item
```

```python
# Impressão das três maiores similaridades por item
for col in similaridade_item_item.columns:
    print(f'  {col}')
    print(similaridade_item_item[similaridade_item_item.index != col][col].nlargest(K))
```
